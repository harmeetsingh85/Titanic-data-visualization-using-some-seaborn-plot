```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset=sns.load_dataset('titanic')
```

```python
dataset.head()
```

```python
sns.distplot(x=dataset['age'],kde=False,bins=30)
```

```python
sns.jointplot(x='age',y='fare',data=dataset,kind='scatter',color='red')
```

```python
sns.rugplot(x='age',data=dataset)
```

```python
sns.kdeplot(dataset['age'])
```

# categorical plots

```python
sns.barplot(x='class',y='survived',data=dataset)
```

```python
sns.boxplot(x='class',y='age',data=dataset,hue='alive')
```

```python
sns.violinplot(x='class',y='age',data=dataset,hue='alive',split=True)
```

```python
sns.swarmplot(x='class',y='age',data=dataset,hue='sex',split='True')
```

```python
sns.stripplot(x='class',y='fare',data=dataset,jitter=True,hue='alive',split=True)
```

```python
sns.countplot(x=dataset['class'])
```
